### Turistskatt p√• overnattinger og dens hensikt
De felles godene vi tar nytte av koster penger. En m√•te for √• finansiere disse kan v√¶re i form av en turistskatt, enten i form av et bestemt bel√∏p eller en prosentandel. En turistskatt p√• overnattinger inneb√¶rer en ekstra moms n√•r du betaler for oppholdet ditt. 

Hensikten med en slik skatt har flere grunner. Selv om det ikke er en enkelt n√¶ringsakt√∏r som st√•r for norske reisem√•ls fellesgoder, s√• har det effekt p√• attraksjonsverdien, og dermed indirekte effekt p√• bedrifters omsetning. Finansiering av vedlikehold, utvide sesongen og unng√• overturisme p√• de mest popul√¶re stedene er noen av begrunnelsene for dens innf√∏ring. 

### Hvor utbredt er dette tiltaket i andre land og byer
Rundt i verden finner vi en del variasjon mellom hvordan dette er implementert, og det siste √•ret har det v√¶rt mange endringer.
